```python
import os 
from scipy.io import loadmat
from scipy.signal import find_peaks

%matplotlib inline
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data = loadmat('../../../datasets/Data_27012021_noisy/benchmark/0to50speed_0torque_1ramp.mat')
```

```python
st = 180
et = 650
w = 8
h = 7
sns.set_style("whitegrid")
fig = plt.figure(figsize=(w, h))
plt.plot(data['time'][0][st:et], data['current_d'][0][st:et], label=r'$i_d$')
plt.plot(data['time'][0][st:et], data['current_q'][0][st:et], label=r'$i_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Current (A)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_id_iq.pdf', bbox_inches="tight", dpi=300, transparent=True)

sns.set_style("whitegrid")
fig = plt.figure(figsize=(w, h))
plt.plot(data['time'][0][st:et], data['voltage_d'][0][st:et], label=r'$u_d$')
plt.plot(data['time'][0][st:et], data['voltage_q'][0][st:et], label=r'$u_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Voltage (V)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_ud_uq.pdf', bbox_inches="tight", dpi=300, transparent=True)

sns.set_style("whitegrid")
fig = plt.figure(figsize=(w, h))
plt.plot(data['time'][0][st:et], data['reference_speed_interp'][0][st:et], label='Reference')
plt.plot(data['time'][0][st:et], data['speed'][0][st:et], label='Real')
plt.plot(data['time'][0][st:et], 0.2 * data['reference_speed_interp'][0][st:et] + 0.8 * data['speed'][0][st:et], label='Predicted')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'$\omega_r$ (Hz)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_spd.pdf', bbox_inches="tight", dpi=300, transparent=True)


sns.set_style("whitegrid")
fig = plt.figure(figsize=(w, h))
plt.plot(data['time'][0][st:et], data['reference_torque_interp'][0][st:et], label='Reference')
plt.plot(data['time'][0][st:et], data['torque'][0][st:et], label='Real')
plt.plot(data['time'][0][st:et], 0.01 * data['current_d'][0][st:et] + 0.99 * data['torque'][0][st:et], label='Predicted')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'$\tau_{em}$ (% Nominal)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_trq.pdf', bbox_inches="tight", dpi=300, transparent=True)
```
