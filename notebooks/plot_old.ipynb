```python
import os 
from scipy.io import loadmat

%matplotlib inline
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
data = loadmat('../../../datasets/Data_27012021_noisy/benchmark/0to50speed_0torque_1ramp.mat')
```

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['current_d'][0][:500], label=r'$i_d$')
plt.plot(data['time'][0][:500], data['current_q'][0][:500], label=r'$i_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Current (A)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_id_iq.pdf', bbox_inches="tight", dpi=300, transparent=True)

sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['voltage_d'][0][:500], label=r'$u_d$')
plt.plot(data['time'][0][:500], data['voltage_q'][0][:500], label=r'$u_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Voltage (V)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_ud_uq.pdf', bbox_inches="tight", dpi=300, transparent=True)
```

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['reference_speed_interp'][0][:500], label='Reference')
plt.plot(data['time'][0][:500], data['speed'][0][:500], label='Real')
plt.plot(data['time'][0][:500], 0.2 * data['reference_speed_interp'][0][:500] + 0.8 * data['speed'][0][:500], label='Predicted')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'$\omega_r$ (Hz)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_spd.pdf', bbox_inches="tight", dpi=300, transparent=True)


sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['reference_torque_interp'][0][:500] / 2.5, label='Reference')
plt.plot(data['time'][0][:500], data['torque'][0][:500] / 2.5, label='Real')
plt.plot(data['time'][0][:500], 0.08 * data['reference_torque_interp'][0][:500] / 2.5+ 0.92 * data['torque'][0][:500] / 2.5, label='Predicted')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'$\tau_{em}$ (% Nominal)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_trq.pdf', bbox_inches="tight", dpi=300, transparent=True)
```

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['noisy_current_d'][0][:500], label=r'$i_d$')
plt.plot(data['time'][0][:500], data['noisy_current_q'][0][:500], label=r'$i_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Current (A)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_sim_noise_id_iq.pdf', bbox_inches="tight", dpi=300, transparent=True)

sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['noisy_voltage_d'][0][:500], label=r'$u_d$')
plt.plot(data['time'][0][:500], data['noisy_voltage_q'][0][:500], label=r'$u_q$')
plt.xlabel('Time (s)', fontsize=30)
plt.ylabel(r'Voltage (V)', fontsize=30)
plt.legend(fontsize=25)
plt.tick_params(axis='both', which='major', labelsize=25)
plt.tick_params(axis='both', which='minor', labelsize=25)
plt.savefig('b1_sim_noise_ud_uq.pdf', bbox_inches="tight", dpi=300, transparent=True)
```

```python
sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['reference_speed_interp'][0][:500], label='Reference')
plt.plot(data['time'][0][:500], data['speed'][0][:500], label='Real')
plt.plot(data['time'][0][:500], 0.5 * data['noisy_current_d'][0][:500] + 0.19 * data['reference_speed_interp'][0][:500] + 0.8 * data['speed'][0][:500], label='Predicted')
plt.xlabel('Time (s)', fontsize=20)
plt.ylabel(r'$\omega_r$ (Hz)', fontsize=20)
plt.legend(fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.tick_params(axis='both', which='minor', labelsize=18)
plt.savefig('b1_sim_noise_spd.pdf', bbox_inches="tight", dpi=300, transparent=True)


sns.set_style("whitegrid")
fig = plt.figure(figsize=(25, 10))
plt.plot(data['time'][0][:500], data['reference_torque_interp'][0][:500], label='Reference')
plt.plot(data['time'][0][:500], data['torque'][0][:500], label='Real')
plt.plot(data['time'][0][:500], 0.5 * data['noisy_voltage_d'][0][:500] + 0.19826261 * data['reference_torque_interp'][0][:500] + 0.8 * data['torque'][0][:500], label='Predicted')
plt.xlabel('Time (s)', fontsize=20)
plt.ylabel(r'$\tau_{em}$ (% Nominal)', fontsize=20)
plt.legend(fontsize=18)
plt.tick_params(axis='both', which='major', labelsize=18)
plt.tick_params(axis='both', which='minor', labelsize=18)
plt.savefig('b1_sim_noise_trq.pdf', bbox_inches="tight", dpi=300, transparent=True)
```
